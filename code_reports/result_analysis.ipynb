```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns

# —— Choose which model suffix you used —— 
ALGO = "ppo"          # for rollout, beam, bfs-fix
ASTAR_ALGO = "ppo"    # for astar

# —— Modern Style Setup —— 
sns.set_theme(
    context='paper',
    style='white',
    palette='deep'
)
plt.rcParams.update({
    'axes.edgecolor':    '#333333',
    'axes.linewidth':    1.0,
    'axes.grid':         True,
    'axes.spines.top':   False,
    'axes.spines.right': False,
    'grid.color':        '#EEEEEE',
    'grid.linewidth':    0.7,
    'legend.frameon':    True,
    'legend.facecolor':  'white',
    'legend.edgecolor':  '#CCCCCC',
    'figure.figsize':    (8, 6),
    'figure.dpi':        150,
    'lines.linewidth':   1.5,
    'lines.marker':      '',
    'font.family':       'sans-serif',
})

# —— Metrics Definition —— 
metrics = ['total_collisions', 'unique_collisions',
           'success_rate', 'total_time_sec', 'makespan']

def load_and_extract(path):
    with open(path, 'rb') as f:
        runs = pickle.load(f)
    data = {m: [] for m in metrics}
    for run in runs:
        rec = run.get('result', run)
        for m in metrics:
            if m == 'unique_collisions':
                v = rec.get('unique_collisions',
                            rec.get('uniques_collisions', np.nan))
            else:
                v = rec.get(m, np.nan)
            data[m].append(v if (m!='success_rate' or v==1.0) else 0.0)
    return {m: np.nanmean(data[m]) for m in metrics}

# —— Main Extraction —— 
agent_counts = np.arange(5, 21)
methods = ['rollout', 'beam', 'bfs-fix', 'astar-ppo', 'cbs', 'icbs']
containers = ['success_rate','total_time_sec','total_collisions','makespan']
avg = {method: {k: [] for k in containers} for method in methods}

# file suffix patterns
suffix = {
    'rollout':   f"_rollout_{ALGO}",
    'beam':      f"_beam_{ALGO}",
    'bfs-fix':   f"_greedy-bfs_{ALGO}",
    'astar-ppo': f"_astar_{ASTAR_ALGO}",
    'cbs':       "_cbs",
    'icbs':      "_icbs",
}

for n in agent_counts:
    results = {}
    for m in methods:
        fname = f"simulation_data/results/simulation_results{suffix[m]}_{n}.pkl"
        if os.path.exists(fname):
            res = load_and_extract(fname)
            if m in ('cbs','icbs'):
                res['total_collisions'] = res.get('unique_collisions', np.nan)
        else:
            res = {dim: np.nan for dim in metrics}
        results[m] = res
    for m in methods:
        for dim in containers:
            avg[m][dim].append(results[m][dim])

# Convert to arrays and mask invalid for log scale
for m in methods:
    for dim in containers:
        arr = np.array(avg[m][dim], dtype=float)
        if dim=='total_collisions' and m in ('cbs','icbs'):
            arr[arr <= 0] = np.nan
        avg[m][dim] = arr

# —— Plotting —— 
fig, axs = plt.subplots(2, 2, constrained_layout=True)
axs = axs.flatten()

specs = [
    ('success_rate',     'Average Success Rate',        'Success Rate'),
    ('total_time_sec',   'Average Total Time (s)',       'Time (s)'),
    ('total_collisions', 'Average Number of Collisions', '# Collisions'),
    ('makespan',         'Average Makespan (steps)',     'Makespan'),
]
labels = {
    'rollout':   'Rollout',
    'beam':      'Beam',
    'bfs-fix':   'BFS-Fix',
    'astar-ppo': 'A*-Fix',
    'cbs':       'CBS',
    'icbs':      'ICBS'
}
linestyles = {
    'rollout':   '-',
    'beam':      ':',
    'bfs-fix':   '--',
    'astar-ppo': '-.',
    'cbs':       '-.',
    'icbs':      ':'
}

for ax, (key, title, ylabel) in zip(axs, specs):
    for m in methods:
        ax.plot(agent_counts, avg[m][key],
                linestyle=linestyles[m],
                label=labels[m])
    ax.set_title(title)
    ax.set_xlabel('Number of Agents')
    ax.set_ylabel(ylabel)
    ax.tick_params(direction='in', top=False, right=False)
    ax.legend(fontsize=6, framealpha=0.5)

# log scale for collisions
axs[2].set_yscale('log')
axs[2].set_ylabel('# Collisions (log scale)')

plt.show()
```

```python
import os
import pickle
import numpy as np

# --- Configuration ---
DATA_DIR   = "simulation_data/results" # Ensure this path is correct

methods_keys = ['bfs-dqn', 'astar-dqn', 'cbs', 'icbs', 'primal', 'scrimp']
method_labels_for_table = ['Our-BFS', 'Our-A*', 'CBS', 'ICBS', 'PRIMAL', 'SCRIMP']
agent_counts_individual = np.arange(5, 21) # For 11x11 grid

suffix = {
    'bfs-dqn':   f"_do_bfs_dqn",
    'astar-dqn': f"_do_astar_dqn",
    'cbs':       "_CBS",
    'icbs':      "_ICBS",
    'primal':    "_primal",
    'scrimp':    "_scrimp",
}

# Added 'episode_len'
table_metric_map_ordered = {
    'success_rate': 'SR',
    'makespan': 'MS',
    'total_collisions': 'CO',
    'total_time_sec': 'T(s)',
    'episode_len': 'EpLen' # Will be used internally and combined for PRIMAL/SCRIMP time
}

# --- Data Loading and Processing Function (UPDATED) ---
def load_and_extract(path, metric_key, is_success_rate_metric, method_file_key):
    try:
        with open(path, 'rb') as f:
            runs_data = pickle.load(f)
    except FileNotFoundError:
        return np.nan
    
    metric_values_for_averaging = []
    if not runs_data: return np.nan

    for run_entry in runs_data:
        result_dict = run_entry.get('result', run_entry)
        if not isinstance(result_dict, dict): continue

        current_run_success_rate = result_dict.get('success_rate', 0.0)
        val_to_append = np.nan 

        if is_success_rate_metric: # Handles 'success_rate'
            metric_values_for_averaging.append(1.0 if current_run_success_rate == 1.0 else 0.0)
            continue 

        # --- Logic for other metrics (MS, CO, T(s), EpLen) ---
        if metric_key == 'total_time_sec':
            # Time (T(s)) is averaged over ALL runs for all methods.
            val_to_append = result_dict.get('total_time_sec', np.nan)
        elif metric_key == 'episode_len':
            # Episode Length handling varies by method
            if method_file_key == 'primal':
                if current_run_success_rate == 1.0:
                    val_to_append = result_dict.get('makespan', np.nan) # EpLen is Makespan for solved PRIMAL
                else:
                    val_to_append = 256.0 # EpLen is 256 for unsolved PRIMAL
            else: # For SCRIMP and other methods, get 'episode_len' directly, average over all runs
                val_to_append = result_dict.get('episode_len', np.nan)
        elif metric_key == 'makespan' or metric_key == 'total_collisions':
            # Makespan (MS) and Collisions (CO) are averaged over SOLVED runs only.
            if current_run_success_rate == 1.0:
                if metric_key == 'total_collisions':
                    # Using the logic from the user-provided script for collisions:
                    if method_file_key in ['bfs-dqn', 'astar-dqn', 'scrimp']:
                        val_to_append = result_dict.get('total_collisions', np.nan)
                    else: # For CBS, ICBS, PRIMAL (if it had collisions for solved)
                        val_to_append = result_dict.get('unique_collisions', np.nan)
                elif metric_key == 'makespan':
                    # This makespan is for methods other than PRIMAL's episode_len use case
                    val_to_append = result_dict.get('makespan', np.nan)
        
        if val_to_append is not None and not np.isnan(val_to_append):
            metric_values_for_averaging.append(float(val_to_append))

    if not metric_values_for_averaging:
        return np.nan
    return np.nanmean(metric_values_for_averaging)

# --- Load average metrics for each individual agent count (5 to 20) ---
avg_metrics_data_individual = {
    method_key: {
        metric_key_in_file: np.full(len(agent_counts_individual), np.nan) # Initialize with NaNs
        for metric_key_in_file in table_metric_map_ordered.keys()
    }
    for method_key in methods_keys
}

print("Loading data for 11x11 grid (agents 5 to 20)...")
for method_idx, method_file_key in enumerate(methods_keys):
    print(f"  Loading for method: {method_labels_for_table[method_idx]} ({method_file_key})")
    if method_file_key not in suffix:
        print(f"    Warning: Suffix for '{method_file_key}' not found. Skipping this method.")
        continue
    for metric_key_in_file in table_metric_map_ordered.keys():
        is_success_rate_metric = (metric_key_in_file == 'success_rate')
        metric_averages_for_method = []
        for num_agents in agent_counts_individual:
            filename = os.path.join(DATA_DIR, f"simulation_results{suffix[method_file_key]}_{num_agents}.pkl")
            avg_val = load_and_extract(filename, metric_key_in_file, is_success_rate_metric, method_file_key)
            metric_averages_for_method.append(avg_val)
        
        processed_averages = np.array(metric_averages_for_method, dtype=float)
        
        if metric_key_in_file == 'total_collisions': 
            processed_averages[processed_averages <= 0] = np.nan
            
        avg_metrics_data_individual[method_file_key][metric_key_in_file] = processed_averages
print("Data loading complete.\n")

# --- Define Agent Groups for 11x11 Table (ordered) ---
agent_groups_11x11_ordered = {
    "5-10": (5, 10),
    "11-15": (11, 15),
    "16-20": (16, 20)
}

# --- Calculate Averages for Agent Groups ---
table_data_11x11 = {
    method_label: {
        metric_abbr: {grp_label: np.nan for grp_label in agent_groups_11x11_ordered}
        for metric_abbr in table_metric_map_ordered.values() # Includes 'EpLen' now
    }
    for method_label in method_labels_for_table
}
min_agent_from_config = agent_counts_individual[0]

for method_idx, method_file_key in enumerate(methods_keys):
    current_method_label = method_labels_for_table[method_idx]
    for metric_key_in_file, table_metric_abbr in table_metric_map_ordered.items():
        if method_file_key not in avg_metrics_data_individual or \
           metric_key_in_file not in avg_metrics_data_individual[method_file_key]:
            print(f"Warning: Pre-loaded data missing for {current_method_label}, metric {metric_key_in_file}. Skipping group calculation.")
            continue
        individual_averages = avg_metrics_data_individual[method_file_key][metric_key_in_file]

        for group_label, (start_agent, end_agent) in agent_groups_11x11_ordered.items():
            start_idx = start_agent - min_agent_from_config
            end_idx = end_agent - min_agent_from_config + 1 

            if start_idx < 0 or end_idx > len(individual_averages) or start_idx >= end_idx :
                # This warning should ideally not trigger if data loading was complete
                print(f"Warning: Agent group {group_label} ({start_agent}-{end_agent}) has invalid slice [{start_idx}:{end_idx}] for {current_method_label}, metric {metric_key_in_file}.")
                table_data_11x11[current_method_label][table_metric_abbr][group_label] = np.nan
                continue
                
            group_values = individual_averages[start_idx:end_idx]
            mean_for_group = np.nanmean(group_values) 
            table_data_11x11[current_method_label][table_metric_abbr][group_label] = mean_for_group

# --- Format and Print Results for 11x11 Table ---
print("--- Results for 11x11 Grid Table ---")
all_methods_output_for_file = []

# UPDATED Data Presentation Note
all_methods_output_for_file.append(f"Data Presentation Note:")
all_methods_output_for_file.append(f"  - Success Rate (SR) is averaged over all trials.")
all_methods_output_for_file.append(f"  - Time (T(s)) and general Episode Length (EpLen) are averaged over ALL trials.")
all_methods_output_for_file.append(f"  - For PRIMAL, EpLen is Makespan for solved instances, 256 for unsolved.")
all_methods_output_for_file.append(f"  - For SCRIMP & PRIMAL, T(s) is shown as T(s) (Avg EpLen).")
all_methods_output_for_file.append(f"  - Makespan (MS) and Collisions (CO) are averaged *only* over successful trials.")
all_methods_output_for_file.append(f"  - A '-' for MS/CO indicates no successful trials with valid data for that metric.")
all_methods_output_for_file.append(f"  - A '-' for T(s)/EpLen indicates no trials provided valid data.\n")

# Create a list of metric abbreviations for the table header, excluding 'EpLen' as it's combined
header_metric_abbrs = [abbr for key, abbr in table_metric_map_ordered.items() if key != 'episode_len']
all_methods_output_for_file.append(f"LaTeX Table Column Structure Hint:")
all_methods_output_for_file.append(f"  Metrics (super-columns): {header_metric_abbrs}")
all_methods_output_for_file.append(f"  Agent Groups per Metric (sub-columns): {list(agent_groups_11x11_ordered.keys())}\n")

for method_idx, method_file_key in enumerate(methods_keys):
    current_method_label = method_labels_for_table[method_idx]
    current_method_console_output_lines = [f"Method: {current_method_label}"]
    current_method_file_output_lines = [f"Method: {current_method_label}"] 
    
    latex_row_cells = [current_method_label] 

    # Iterate through metrics for display, handling T(s) and EpLen specially
    for metric_key_in_file_display, metric_abbr_display in table_metric_map_ordered.items():
        if metric_key_in_file_display == 'episode_len': # Skip EpLen as a separate column/output section
            continue

        current_method_console_output_lines.append(f"  {metric_abbr_display}:")
        current_method_file_output_lines.append(f"  {metric_abbr_display}:")
        
        metric_group_values_for_latex_row = [] 

        for group_label in agent_groups_11x11_ordered:
            raw_value = table_data_11x11[current_method_label][metric_abbr_display].get(group_label, np.nan)
            
            human_readable_formatted_value = "-" 
            if not np.isnan(raw_value):
                if metric_abbr_display == 'SR':
                    human_readable_formatted_value = f"{raw_value * 100:.2f}%"
                elif metric_abbr_display == 'CO':
                    human_readable_formatted_value = f"{raw_value:.0f}"
                else: # MS, T(s)
                    human_readable_formatted_value = f"{raw_value:.1f}"
            
            # Special handling for SCRIMP and PRIMAL Time to include Episode Length
            if metric_key_in_file_display == 'total_time_sec' and \
               (method_file_key == 'scrimp' or method_file_key == 'primal'):
                # Fetch the corresponding average episode length for this group
                avg_ep_len_for_group = table_data_11x11[current_method_label]['EpLen'].get(group_label, np.nan)
                if not np.isnan(avg_ep_len_for_group):
                    formatted_ep_len = f"{avg_ep_len_for_group:.0f}" # Format EpLen as integer
                    if human_readable_formatted_value != "-":
                        human_readable_formatted_value += f" ({formatted_ep_len})"
                    else: # If time itself is NaN but ep_len exists
                        human_readable_formatted_value = f"- ({formatted_ep_len})"
            
            value_for_console_and_latex_cell = human_readable_formatted_value
            if metric_abbr_display == 'SR' and not np.isnan(raw_value): 
                value_for_console_and_latex_cell = human_readable_formatted_value.replace('%', '\\%') 
            
            current_method_console_output_lines.append(f"    {group_label}: {value_for_console_and_latex_cell}")
            current_method_file_output_lines.append(f"    {group_label}: {human_readable_formatted_value}") 
            metric_group_values_for_latex_row.append(value_for_console_and_latex_cell)
            
        latex_row_cells.extend(metric_group_values_for_latex_row)

    latex_row_string = " & ".join(latex_row_cells) + " \\\\"
    current_method_console_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    current_method_file_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    
    for line in current_method_console_output_lines:
        print(line)
    print("-" * 30)
    
    all_methods_output_for_file.append("\n".join(current_method_file_output_lines))

# # --- Save results to a text file for easy sharing ---
# output_filename = "table_data_11x11_with_eplen_details.txt"
# with open(output_filename, "w") as f:
#     f.write("--- Simulation Data for 11x11 LaTeX Table (with EpLen details for SCRIMP & PRIMAL) ---\n\n")
#     f.write("\n\n".join(all_methods_output_for_file)) 

# print(f"\nFormatted results saved to: {output_filename}")
```

```python
import os
import pickle
import numpy as np

# --- Configuration ---
DATA_DIR   = "simulation_data/results_model_comp" # Ensure this path is correct

# Methods for this specific table data generation
methods_keys_for_table = ['bfs-dqn', 'bfs-ppo', 'astar-dqn', 'astar-ppo']
# Corresponding labels for the table output
method_labels_for_table_output = ['BFS-DQN', 'BFS-PPO', 'A*-DQN', 'A*-PPO']

# Agent counts for which data is loaded (individual agent numbers)
agent_counts_individual = np.arange(5, 21) # Corresponds to agents 5, 6, ..., 20

# Suffixes for filenames
suffix = {
    'bfs-dqn':   f"_do_bfs_dqn",
    'bfs-ppo':   f"_do_bfs_ppo",
    'astar-dqn': f"_do_astar_dqn",
    'astar-ppo': f"_do_astar_ppo",
}

# Mapping internal metric keys from files to the table's desired abbreviations
# The order here will also define the order in the output for metrics.
table_metric_map_ordered = {
    'success_rate': 'SR',
    'makespan': 'MS',
    'total_collisions': 'CO',
    'total_time_sec': 'T(s)'
}

# --- Data Loading and Processing Function (Consistent with previous table script) ---
def load_and_extract(path, metric_key, is_success_rate_metric, method_file_key):
    """
    Load data from a pickle file and extract the mean of a specified metric.
    - path: Path to the pickle file.
    - metric_key: The key for the metric in the data dictionary.
    - is_success_rate_metric: Boolean, True if the metric is 'success_rate'.
    - method_file_key: The key for the method, used for specific logic.
    """
    try:
        with open(path, 'rb') as f:
            runs_data = pickle.load(f)
    except FileNotFoundError:
        return np.nan
    
    metric_values_for_averaging = []
    if not runs_data: return np.nan

    for run_entry in runs_data:
        result_dict = run_entry.get('result', run_entry)
        if not isinstance(result_dict, dict): continue

        current_run_success_rate = result_dict.get('success_rate', 0.0)
        val_to_append = np.nan 

        if is_success_rate_metric:
            metric_values_for_averaging.append(1.0 if current_run_success_rate == 1.0 else 0.0)
            continue 

        if metric_key == 'total_time_sec': # Time is averaged over ALL runs
            val_to_append = result_dict.get('total_time_sec', np.nan)
        elif metric_key == 'makespan' or metric_key == 'total_collisions': # MS and CO over SOLVED runs
            if current_run_success_rate == 1.0:
                if metric_key == 'total_collisions':
                    # DQN/PPO methods likely use 'total_collisions'
                    if method_file_key in ['bfs-dqn', 'bfs-ppo', 'astar-dqn', 'astar-ppo']:
                        val_to_append = result_dict.get('total_collisions', np.nan)
                    else: # For other methods like CBS, ICBS (if they were included)
                        val_to_append = result_dict.get('unique_collisions', np.nan)
                elif metric_key == 'makespan':
                    val_to_append = result_dict.get('makespan', np.nan)
        
        if val_to_append is not None and not np.isnan(val_to_append):
            metric_values_for_averaging.append(float(val_to_append))

    if not metric_values_for_averaging:
        return np.nan
    return np.nanmean(metric_values_for_averaging)

# --- Load average metrics for each individual agent count ---
avg_metrics_data_individual = {method_key: {} for method_key in methods_keys_for_table}
print("Loading data for individual agent counts (for table)...")
for method_idx, method_file_key in enumerate(methods_keys_for_table):
    print(f"  Loading for method: {method_labels_for_table_output[method_idx]} ({method_file_key})")
    for metric_key_in_file in table_metric_map_ordered.keys():
        is_success_rate_metric = (metric_key_in_file == 'success_rate')
        metric_averages_for_method = []
        for num_agents in agent_counts_individual:
            filename_suffix_key = method_file_key # Key to look up in suffix dict
            if filename_suffix_key not in suffix:
                print(f"    Warning: Suffix for {filename_suffix_key} not found. Cannot generate filename for {num_agents} agents.")
                avg_val = np.nan
            else:
                filename = os.path.join(DATA_DIR, f"simulation_results{suffix[filename_suffix_key]}_{num_agents}.pkl")
                avg_val = load_and_extract(filename, metric_key_in_file, is_success_rate_metric, method_file_key)
            metric_averages_for_method.append(avg_val)
        
        processed_averages = np.array(metric_averages_for_method, dtype=float)
        
        if metric_key_in_file == 'total_collisions': 
            processed_averages[processed_averages <= 0] = np.nan
            
        avg_metrics_data_individual[method_file_key][metric_key_in_file] = processed_averages
print("Data loading for table complete.\n")

# --- Define Agent Groups for Table (e.g., 11x11 style) ---
agent_groups_for_table_ordered = {
    "5-10": (5, 10),
    "11-15": (11, 15),
    "16-20": (16, 20)
}

# --- Calculate Averages for Agent Groups ---
table_data_new_methods = {
    method_label: {
        metric_abbr: {grp_label: np.nan for grp_label in agent_groups_for_table_ordered}
        for metric_abbr in table_metric_map_ordered.values()
    }
    for method_label in method_labels_for_table_output
}
min_agent_from_config = agent_counts_individual[0]

for method_idx, method_file_key in enumerate(methods_keys_for_table):
    current_method_label = method_labels_for_table_output[method_idx]
    for metric_key_in_file, table_metric_abbr in table_metric_map_ordered.items():
        # Ensure the metric_key_in_file was loaded for the method
        if metric_key_in_file not in avg_metrics_data_individual[method_file_key]:
             print(f"Warning: Metric key {metric_key_in_file} not found in loaded data for {method_file_key}. Skipping for table.")
             continue
        individual_averages = avg_metrics_data_individual[method_file_key][metric_key_in_file]

        for group_label, (start_agent, end_agent) in agent_groups_for_table_ordered.items():
            start_idx = start_agent - min_agent_from_config
            end_idx = end_agent - min_agent_from_config + 1 

            if start_idx < 0 or end_idx > len(individual_averages) or start_idx >= end_idx :
                print(f"Warning: Agent group {group_label} ({start_agent}-{end_agent}) has invalid slice [{start_idx}:{end_idx}] for loaded data.")
                table_data_new_methods[current_method_label][table_metric_abbr][group_label] = np.nan
                continue
                
            group_values = individual_averages[start_idx:end_idx]
            mean_for_group = np.nanmean(group_values) 
            table_data_new_methods[current_method_label][table_metric_abbr][group_label] = mean_for_group

# --- Format and Print Results for New Methods Table ---
print("--- Results for New Methods Table (BFS-DQN/PPO, A*-DQN/PPO) ---")
all_methods_output_for_file = []

all_methods_output_for_file.append(f"Data Presentation Note:")
all_methods_output_for_file.append(f"  - Success Rate (SR) is averaged over all trials.")
all_methods_output_for_file.append(f"  - Time (T(s)) is averaged over ALL trials (reflecting timeouts).")
all_methods_output_for_file.append(f"  - Makespan (MS) and Collisions (CO) are averaged *only* over successful trials.")
all_methods_output_for_file.append(f"  - A '-' for MS/CO indicates no successful trials in group with valid data.")
all_methods_output_for_file.append(f"  - A '-' for T(s) indicates no trials in group provided valid time data.\n")

all_methods_output_for_file.append(f"LaTeX Table Column Structure Hint (Example for 11x11 style):")
all_methods_output_for_file.append(f"  Metrics (super-columns): {list(table_metric_map_ordered.values())}")
all_methods_output_for_file.append(f"  Agent Groups per Metric (sub-columns): {list(agent_groups_for_table_ordered.keys())}\n")

for method_label in method_labels_for_table_output:
    current_method_console_output_lines = [f"Method: {method_label}"]
    current_method_file_output_lines = [f"Method: {method_label}"] 
    
    latex_row_cells = [method_label] 

    for metric_abbr in table_metric_map_ordered.values(): 
        current_method_console_output_lines.append(f"  {metric_abbr}:")
        current_method_file_output_lines.append(f"  {metric_abbr}:")
        
        metric_group_values_for_latex_row = [] 

        for group_label in agent_groups_for_table_ordered:
            raw_value = table_data_new_methods[method_label][metric_abbr][group_label]
            
            human_readable_formatted_value = "-" 
            if not np.isnan(raw_value):
                if metric_abbr == 'SR':
                    human_readable_formatted_value = f"{raw_value * 100:.2f}%"
                elif metric_abbr == 'CO':
                    human_readable_formatted_value = f"{raw_value:.0f}"
                else: # MS, T(s)
                    human_readable_formatted_value = f"{raw_value:.1f}"
            
            value_for_console_and_latex_cell = human_readable_formatted_value
            if metric_abbr == 'SR' and not np.isnan(raw_value): 
                value_for_console_and_latex_cell = human_readable_formatted_value.replace('%', '\\%') 
            
            current_method_console_output_lines.append(f"    {group_label}: {value_for_console_and_latex_cell}")
            current_method_file_output_lines.append(f"    {group_label}: {human_readable_formatted_value}") 
            metric_group_values_for_latex_row.append(value_for_console_and_latex_cell)
            
        latex_row_cells.extend(metric_group_values_for_latex_row)

    latex_row_string = " & ".join(latex_row_cells) + " \\\\"
    current_method_console_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    current_method_file_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    
    for line in current_method_console_output_lines:
        print(line)
    print("-" * 30)
    
    all_methods_output_for_file.append("\n".join(current_method_file_output_lines))

# --- Save results to a text file ---
output_filename = "table_data_new_methods_results.txt"
with open(output_filename, "w") as f:
    f.write("--- Simulation Data for New Methods Table (BFS-DQN/PPO, A*-DQN/PPO) ---\n")
    f.write("--- Time (T(s)) averaged over ALL trials ---\n\n")
    f.write("\n\n".join(all_methods_output_for_file)) 

print(f"\nFormatted results for new methods table saved to: {output_filename}")

```

# 21x21 results

```python
import os
import pickle
import numpy as np

# --- Configuration ---
DATA_DIR   = "simulation_data/results" # Ensure this path is correct

# Methods for this specific table data generation
# Now includes all methods for which suffixes are defined
methods_keys_for_table = ['bfs-dqn', 'astar-dqn', 'bfs-ppo', 'astar-ppo', 'scrimp', 'primal', 'cbs', 'icbs']
# Corresponding labels for the table output
method_labels_for_table_output = ['Our-BFS-DQN', 'Our-A*-DQN', 'Our-BFS-PPO', 'Our-A*-PPO', 'SCRIMP', 'PRIMAL', 'CBS', 'ICBS']

# Agent counts to process for the 21x21 grid
agent_counts_to_process = [32, 64, 96]

# Suffixes for filenames - ensure these match your actual filenames
suffix = {
    'bfs-dqn':   "_do_bfs_dqn",
    'astar-dqn': "_do_astar_dqn",
    'bfs-ppo':   "_do_bfs_ppo",
    'astar-ppo': "_do_astar_ppo",
    'scrimp':    "_scrimp",
    'primal':    "_primal",
    'cbs':       "_CBS",
    'icbs':      "_ICBS",
}

# Mapping internal metric keys from files to the table's desired abbreviations
table_metric_map_ordered = {
    'success_rate': 'SR',
    'makespan': 'MS',
    'total_collisions': 'CO',
    'total_time_sec': 'T(s)',
    'episode_len': 'EpLen' # Internal key for episode length
}

# --- Data Loading and Processing Function (UPDATED) ---
def load_and_extract(path, metric_key, is_success_rate_metric, method_file_key):
    """
    Load data from a pickle file and extract the mean of a specified metric.
    - path: Path to the pickle file.
    - metric_key: The key for the metric in the data dictionary.
    - is_success_rate_metric: Boolean, True if the metric is 'success_rate'.
    - method_file_key: The key for the method, used for specific logic.
    """
    try:
        with open(path, 'rb') as f:
            runs_data = pickle.load(f)
    except FileNotFoundError:
        # print(f"Warning: File not found {path}")
        return np.nan
    
    metric_values_for_averaging = []
    if not runs_data: return np.nan

    for run_entry in runs_data:
        result_dict = run_entry.get('result', run_entry)
        if not isinstance(result_dict, dict): continue

        current_run_success_rate = result_dict.get('success_rate', 0.0)
        val_to_append = np.nan 

        if is_success_rate_metric: # This is 'success_rate'
            metric_values_for_averaging.append(1.0 if current_run_success_rate == 1.0 else 0.0)
            continue 

        # For other metrics
        if metric_key == 'total_time_sec': # Time is averaged over ALL runs
            val_to_append = result_dict.get('total_time_sec', np.nan)
        elif metric_key == 'episode_len': # Episode Length handling
            if method_file_key == 'primal':
                if current_run_success_rate == 1.0:
                    val_to_append = result_dict.get('makespan', np.nan) # EpLen is Makespan for solved PRIMAL
                else:
                    val_to_append = 512.0 # EpLen is 512 for unsolved PRIMAL (21x21 grid assumption)
            else: # For SCRIMP, CBS, ICBS, and other methods (BFS-DQN, A*-DQN, etc.)
                val_to_append = result_dict.get('episode_len', np.nan) # Averaged over ALL runs
        elif metric_key == 'makespan' or metric_key == 'total_collisions': # MS and CO over SOLVED runs
            if current_run_success_rate == 1.0:
                if metric_key == 'total_collisions':
                    # Logic for collisions based on method type
                    if method_file_key in ['bfs-dqn', 'astar-dqn', 'bfs-ppo', 'astar-ppo', 'scrimp']:
                        val_to_append = result_dict.get('total_collisions', np.nan)
                    else: # For PRIMAL, CBS, ICBS if they report 'unique_collisions'
                        val_to_append = result_dict.get('unique_collisions', np.nan)
                elif metric_key == 'makespan':
                    # This is the general makespan, not PRIMAL's episode_len use case for makespan
                    val_to_append = result_dict.get('makespan', np.nan)
        
        if val_to_append is not None and not np.isnan(val_to_append):
            metric_values_for_averaging.append(float(val_to_append))

    if not metric_values_for_averaging:
        return np.nan
    return np.nanmean(metric_values_for_averaging)

# --- Load average metrics for each specified agent count ---
results_data_21x21 = {
    method_key: {
        metric_key_in_file: {agent_count: np.nan for agent_count in agent_counts_to_process}
        for metric_key_in_file in table_metric_map_ordered.keys() 
    }
    for method_key in methods_keys_for_table # Use the full list of methods
}

print(f"Loading data for 21x21 grid (Agents: {agent_counts_to_process})...")
for method_idx, method_file_key in enumerate(methods_keys_for_table):
    print(f"  Loading for method: {method_labels_for_table_output[method_idx]} ({method_file_key})")
    if method_file_key not in suffix:
        print(f"    Warning: Suffix for '{method_file_key}' not found. Skipping this method.")
        # Initialize with NaNs if suffix is missing to prevent key errors later
        for metric_key_in_file in table_metric_map_ordered.keys():
            for num_agents in agent_counts_to_process:
                results_data_21x21[method_file_key][metric_key_in_file][num_agents] = np.nan
        continue

    for metric_key_in_file in table_metric_map_ordered.keys(): 
        is_success_rate_metric = (metric_key_in_file == 'success_rate')
        for num_agents in agent_counts_to_process:
            filename = os.path.join(DATA_DIR, f"simulation_results{suffix[method_file_key]}_{num_agents}.pkl")
            avg_val = load_and_extract(filename, metric_key_in_file, is_success_rate_metric, method_file_key)
            
            if metric_key_in_file == 'total_collisions' and avg_val is not None and avg_val <= 0:
                avg_val = np.nan 
            
            results_data_21x21[method_file_key][metric_key_in_file][num_agents] = avg_val
print("Data loading complete.\n")


# --- Format and Print Results for 21x21 Table ---
print(f"--- Results for 21x21 Grid Table (Methods: {', '.join(method_labels_for_table_output)}; Agents: {agent_counts_to_process}) ---")
all_methods_output_for_file = []

# Data Presentation Note
all_methods_output_for_file.append(f"Data Presentation Note:")
all_methods_output_for_file.append(f"  - Grid: 21x21")
all_methods_output_for_file.append(f"  - Agents: {agent_counts_to_process}")
all_methods_output_for_file.append(f"  - Success Rate (SR) is averaged over all trials.")
all_methods_output_for_file.append(f"  - Time (T(s)) and general Episode Length (EpLen) are averaged over ALL trials.")
all_methods_output_for_file.append(f"  - For PRIMAL, EpLen is Makespan for solved instances, 512 for unsolved (21x21 grid).")
all_methods_output_for_file.append(f"  - For SCRIMP & PRIMAL, T(s) is shown as T(s) (Avg EpLen).")
all_methods_output_for_file.append(f"  - Makespan (MS) and Collisions (CO) are averaged *only* over successful trials.")
all_methods_output_for_file.append(f"  - A '-' for MS/CO indicates no successful trials with valid data for that metric.")
all_methods_output_for_file.append(f"  - A '-' for T(s)/EpLen indicates no trials provided valid data.\n")

header_metric_abbrs = [abbr for key, abbr in table_metric_map_ordered.items() if key != 'episode_len']
all_methods_output_for_file.append(f"LaTeX Table Column Structure Hint (Example for 21x21 style):")
all_methods_output_for_file.append(f"  Metrics (super-columns): {header_metric_abbrs}")
all_methods_output_for_file.append(f"  Agent Counts per Metric (sub-columns): {agent_counts_to_process}\n")


for method_idx, method_file_key in enumerate(methods_keys_for_table):
    current_method_label = method_labels_for_table_output[method_idx]
    current_method_console_output_lines = [f"Method: {current_method_label}"]
    current_method_file_output_lines = [f"Method: {current_method_label}"]
    
    latex_row_cells = [current_method_label]

    for metric_key_in_file_display, metric_abbr_display in table_metric_map_ordered.items():
        if metric_key_in_file_display == 'episode_len': 
            continue

        current_method_console_output_lines.append(f"  {metric_abbr_display}:")
        current_method_file_output_lines.append(f"  {metric_abbr_display}:")
        
        metric_agent_count_values_for_latex_row = []

        for num_agents in agent_counts_to_process:
            raw_value = np.nan # Default to handle cases where method_file_key might be missing after suffix check
            if method_file_key in results_data_21x21 and \
               metric_key_in_file_display in results_data_21x21[method_file_key]:
                raw_value = results_data_21x21[method_file_key][metric_key_in_file_display].get(num_agents, np.nan)
            else:
                # This case should be rare if suffix check is passed and results_data_21x21 is initialized for all methods_keys_for_table
                print(f"    Warning: Data for {current_method_label} - {metric_abbr_display} - {num_agents} Agents not found in results_data_21x21. Using NaN.")


            human_readable_formatted_value = "-" 
            if not np.isnan(raw_value):
                if metric_abbr_display == 'SR':
                    human_readable_formatted_value = f"{raw_value * 100:.2f}%"
                elif metric_abbr_display == 'CO':
                    human_readable_formatted_value = f"{raw_value:.0f}"
                else: # MS, T(s)
                    human_readable_formatted_value = f"{raw_value:.1f}"
            
            # Special handling for SCRIMP and PRIMAL Time to include Episode Length
            if metric_key_in_file_display == 'total_time_sec' and \
               (method_file_key == 'scrimp' or method_file_key == 'primal'):
                avg_ep_len_for_agent_count = np.nan
                if method_file_key in results_data_21x21 and \
                   'episode_len' in results_data_21x21[method_file_key]:
                     avg_ep_len_for_agent_count = results_data_21x21[method_file_key]['episode_len'].get(num_agents, np.nan)

                if not np.isnan(avg_ep_len_for_agent_count):
                    formatted_ep_len = f"{avg_ep_len_for_agent_count:.0f}" 
                    if human_readable_formatted_value != "-":
                        human_readable_formatted_value += f" ({formatted_ep_len})"
                    else: 
                        human_readable_formatted_value = f"- ({formatted_ep_len})"
            
            value_for_console_and_latex_cell = human_readable_formatted_value
            if metric_abbr_display == 'SR' and not np.isnan(raw_value): 
                value_for_console_and_latex_cell = human_readable_formatted_value.replace('%', '\\%')
            
            current_method_console_output_lines.append(f"    {num_agents} Agents: {value_for_console_and_latex_cell}")
            current_method_file_output_lines.append(f"    {num_agents} Agents: {human_readable_formatted_value}") 
            metric_agent_count_values_for_latex_row.append(value_for_console_and_latex_cell)
            
        latex_row_cells.extend(metric_agent_count_values_for_latex_row)

    latex_row_string = " & ".join(latex_row_cells) + " \\\\"
    current_method_console_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    current_method_file_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    
    for line in current_method_console_output_lines:
        print(line)
    print("-" * 30)
    
    all_methods_output_for_file.append("\n".join(current_method_file_output_lines))

# # --- Save results to a text file ---
# output_filename = "table_data_21x21_all_methods_32_64_96_agents_eplen512_v2.txt" # Updated filename
# with open(output_filename, "w") as f:
#     f.write(f"--- Simulation Data for 21x21 Grid (All Methods: {agent_counts_to_process} Agents) ---\n")
#     f.write("--- Time (T(s)) and EpLen averaged over ALL trials; EpLen details for PRIMAL/SCRIMP in T(s) ---\n\n")
#     f.write("\n\n".join(all_methods_output_for_file)) 

# print(f"\nFormatted results saved to: {output_filename}")
```

```python
import os
import pickle
import numpy as np

# --- Configuration ---
# Define the two data directories
DATA_DIR_WR1 = "simulation_data/maps_25x25/results"  # For _wr files
DATA_DIR_WR2 = "simulation_data/maps_25x25_2/results" # For _wr2 files
DATA_DIRS_TO_PROCESS = [DATA_DIR_WR1, DATA_DIR_WR2]

# Methods for this specific table data generation
methods_keys_for_table = ['bfs-dqn', 'astar-dqn', 'bfs-ppo', 'astar-ppo', 'scrimp', 'primal', 'cbs', 'icbs']
# Corresponding labels for the table output
method_labels_for_table_output = ['Our-BFS-DQN', 'Our-A*-DQN', 'Our-BFS-PPO', 'Our-A*-PPO', 'SCRIMP', 'PRIMAL', 'CBS', 'ICBS']

# Agent counts to process
agent_counts_to_process = [32, 64, 96]

# Base Suffixes for filenames (generic part, _wr or _wr2 will be prepended or appended)
base_suffix = {
    'bfs-dqn':   "_do_bfs_dqn",
    'astar-dqn': "_do_astar_dqn",
    'bfs-ppo':   "_do_bfs_ppo",   # Assumed generic (no _wr/_wr2 variants)
    'astar-ppo': "_do_astar_ppo", # Assumed generic
    'scrimp':    "_scrimp",       # _wr or _wr2 will be appended
    'primal':    "_primal",       # _wr or _wr2 will be appended
    'cbs':       "_CBS",
    'icbs':      "_ICBS",
}

# Mapping internal metric keys from files to the table's desired abbreviations
table_metric_map_ordered = {
    'success_rate': 'SR',
    'makespan': 'MS',
    'total_collisions': 'CO',
    'total_time_sec': 'T(s)',
    'episode_len': 'EpLen' 
}

# --- Data Loading and Processing Function (MODIFIED) ---
def load_and_extract_raw_values(path, metric_key, is_success_rate_metric, method_file_key):
    """
    Load data from a pickle file and extract a list of raw metric values.
    - path: Path to the pickle file.
    - metric_key: The key for the metric in the data dictionary.
    - is_success_rate_metric: Boolean, True if the metric is 'success_rate'.
    - method_file_key: The key for the method, used for specific logic.
    """
    try:
        with open(path, 'rb') as f:
            runs_data = pickle.load(f)
    except FileNotFoundError:
        # print(f"Info: File not found {path}") 
        return [] 
    
    raw_metric_values = []
    if not runs_data: return []

    for run_entry in runs_data:
        result_dict = run_entry.get('result', run_entry)
        if not isinstance(result_dict, dict): continue

        current_run_success_rate = result_dict.get('success_rate', 0.0)
        val_to_append = np.nan 

        if is_success_rate_metric:
            raw_metric_values.append(1.0 if current_run_success_rate == 1.0 else 0.0)
            continue 

        if metric_key == 'total_time_sec':
            val_to_append = result_dict.get('total_time_sec', np.nan)
        elif metric_key == 'episode_len':
            if method_file_key == 'primal':
                if current_run_success_rate == 1.0:
                    val_to_append = result_dict.get('makespan', np.nan) 
                else:
                    val_to_append = 512.0 # Step limit for 25x25 warehouse
            else: 
                val_to_append = result_dict.get('episode_len', np.nan)
        elif metric_key == 'makespan' or metric_key == 'total_collisions':
            if current_run_success_rate == 1.0:
                if metric_key == 'total_collisions':
                    if method_file_key in ['bfs-dqn', 'astar-dqn', 'bfs-ppo', 'astar-ppo', 'scrimp']:
                        val_to_append = result_dict.get('total_collisions', np.nan)
                    else: 
                        val_to_append = result_dict.get('unique_collisions', np.nan)
                elif metric_key == 'makespan':
                    val_to_append = result_dict.get('makespan', np.nan)
        
        if val_to_append is not None and not np.isnan(val_to_append):
            raw_metric_values.append(float(val_to_append))

    return raw_metric_values

# --- Load average metrics for each specified agent count ---
# Initialize results structure
results_data_25x25 = {
    method_key: {
        metric_key_in_file: {agent_count: np.nan for agent_count in agent_counts_to_process}
        for metric_key_in_file in table_metric_map_ordered.keys() 
    }
    for method_key in methods_keys_for_table
}

print(f"Loading data for 25x25 Warehouse grid (Agents: {agent_counts_to_process})...")
for method_idx, method_file_key in enumerate(methods_keys_for_table):
    print(f"  Loading for method: {method_labels_for_table_output[method_idx]} ({method_file_key})")
    
    current_base_suffix_for_method = base_suffix.get(method_file_key)
    if current_base_suffix_for_method is None:
        print(f"    Warning: Base suffix for '{method_file_key}' not defined. Skipping this method.")
        for metric_key_in_file in table_metric_map_ordered.keys():
            for num_agents in agent_counts_to_process:
                results_data_25x25[method_file_key][metric_key_in_file][num_agents] = np.nan
        continue

    for metric_key_in_file in table_metric_map_ordered.keys(): 
        is_success_rate_metric = (metric_key_in_file == 'success_rate')
        for num_agents in agent_counts_to_process:
            combined_raw_values_for_metric = []
            
            for data_dir in DATA_DIRS_TO_PROCESS:
                dir_specific_variant_marker = ""
                if data_dir == DATA_DIR_WR1:
                    dir_specific_variant_marker = "_wr"
                elif data_dir == DATA_DIR_WR2:
                    dir_specific_variant_marker = "_wr2"
                
                filename_core_part = ""
                if method_file_key in ['scrimp', 'primal']:
                    # Append _wr or _wr2: e.g., _scrimp_wr or _scrimp_wr2
                    filename_core_part = f"{current_base_suffix_for_method}{dir_specific_variant_marker}"
                elif method_file_key in ['bfs-ppo', 'astar-ppo']:
                    # PPO methods use their generic base suffix, no _wr or _wr2 marker
                    filename_core_part = f"{current_base_suffix_for_method}"
                else: 
                    # For bfs-dqn, astar-dqn, cbs, icbs: Prepend _wr or _wr2
                    filename_core_part = f"{dir_specific_variant_marker}{current_base_suffix_for_method}"
                
                filename = os.path.join(data_dir, f"simulation_results{filename_core_part}_{num_agents}.pkl")
                
                raw_values_from_file = load_and_extract_raw_values(filename, metric_key_in_file, is_success_rate_metric, method_file_key)
                combined_raw_values_for_metric.extend(raw_values_from_file)
            
            if not combined_raw_values_for_metric:
                avg_val = np.nan
            else:
                avg_val = np.nanmean(combined_raw_values_for_metric)
            
            if metric_key_in_file == 'total_collisions':
                # Ensure sr_for_this_config is fetched correctly before use
                sr_for_this_config = results_data_25x25.get(method_file_key, {}).get('success_rate', {}).get(num_agents, 0)
                if sr_for_this_config == 0 and (np.isnan(avg_val) or avg_val == 0):
                    pass 
                elif avg_val is not None and avg_val <= 0 and sr_for_this_config > 0 and method_file_key not in ['scrimp']: 
                    avg_val = np.nan 
            
            results_data_25x25[method_file_key][metric_key_in_file][num_agents] = avg_val
print("Data loading complete.\n")


# --- Format and Print Results for 25x25 Warehouse Table ---
print(f"--- Results for 25x25 Warehouse Grid Table (Methods: {', '.join(method_labels_for_table_output)}; Agents: {agent_counts_to_process}) ---")
all_methods_output_for_file = []

# Data Presentation Note
all_methods_output_for_file.append(f"Data Presentation Note (25x25 Warehouse):")
all_methods_output_for_file.append(f"  - Grid: 25x25 Warehouse")
all_methods_output_for_file.append(f"  - Agents: {agent_counts_to_process}")
all_methods_output_for_file.append(f"  - Results averaged over up to 10 instances (from _wr and _wr2 files where available).")
all_methods_output_for_file.append(f"  - SR is averaged over all trials.")
all_methods_output_for_file.append(f"  - T(s) and EpLen are averaged over ALL trials.")
all_methods_output_for_file.append(f"  - For PRIMAL, EpLen is Makespan for solved, 512 for unsolved (25x25).")
all_methods_output_for_file.append(f"  - For SCRIMP & PRIMAL, T(s) is T(s) (Avg EpLen).")
all_methods_output_for_file.append(f"  - MS and CO are averaged *only* over successful trials.")
all_methods_output_for_file.append(f"  - A '-' for MS/CO implies no successful trials with valid data.")
all_methods_output_for_file.append(f"  - A '-' for T(s)/EpLen implies no trials provided valid data.\n")

header_metric_abbrs = [abbr for key, abbr in table_metric_map_ordered.items() if key != 'episode_len']
all_methods_output_for_file.append(f"LaTeX Table Column Structure Hint:")
all_methods_output_for_file.append(f"  Metrics (super-columns): {header_metric_abbrs}")
all_methods_output_for_file.append(f"  Agent Counts per Metric (sub-columns): {agent_counts_to_process}\n")


for method_idx, method_file_key in enumerate(methods_keys_for_table):
    current_method_label = method_labels_for_table_output[method_idx]
    current_method_console_output_lines = [f"Method: {current_method_label}"]
    current_method_file_output_lines = [f"Method: {current_method_label}"]
    
    latex_row_cells = [current_method_label]

    for metric_key_in_file_display, metric_abbr_display in table_metric_map_ordered.items():
        if metric_key_in_file_display == 'episode_len': 
            continue

        current_method_console_output_lines.append(f"  {metric_abbr_display}:")
        current_method_file_output_lines.append(f"  {metric_abbr_display}:")
        
        metric_agent_count_values_for_latex_row = []

        for num_agents in agent_counts_to_process:
            raw_value = results_data_25x25[method_file_key][metric_key_in_file_display].get(num_agents, np.nan)
            
            human_readable_formatted_value = "-" 
            if not np.isnan(raw_value):
                if metric_abbr_display == 'SR':
                    human_readable_formatted_value = f"{raw_value * 100:.2f}%"
                elif metric_abbr_display == 'CO':
                    sr_val_for_co_check = results_data_25x25[method_file_key]['success_rate'].get(num_agents, 0)
                    if method_file_key == 'scrimp' and raw_value == 0 and sr_val_for_co_check > 0:
                        human_readable_formatted_value = f"{raw_value:.0f}"
                    elif raw_value == 0 and sr_val_for_co_check > 0 and method_file_key != 'scrimp':
                        human_readable_formatted_value = f"{raw_value:.0f}" 
                    elif raw_value > 0 :
                         human_readable_formatted_value = f"{raw_value:.0f}"
                    elif raw_value == 0 and sr_val_for_co_check == 0: # If SR is 0, CO is effectively '-'
                        human_readable_formatted_value = "-"
                else: # MS, T(s)
                    human_readable_formatted_value = f"{raw_value:.1f}"
            
            if metric_key_in_file_display == 'total_time_sec' and \
               (method_file_key == 'scrimp' or method_file_key == 'primal'):
                avg_ep_len_for_agent_count = results_data_25x25[method_file_key]['episode_len'].get(num_agents, np.nan)

                if not np.isnan(avg_ep_len_for_agent_count):
                    formatted_ep_len = f"{avg_ep_len_for_agent_count:.0f}" 
                    if human_readable_formatted_value != "-":
                        human_readable_formatted_value += f" ({formatted_ep_len})"
                    else: 
                        human_readable_formatted_value = f"- ({formatted_ep_len})" 
            
            value_for_latex_cell = human_readable_formatted_value
            if metric_abbr_display == 'SR' and not np.isnan(raw_value): 
                value_for_latex_cell = human_readable_formatted_value.replace('%', '\\%')
            
            current_method_console_output_lines.append(f"    {num_agents} Agents: {human_readable_formatted_value}") 
            current_method_file_output_lines.append(f"    {num_agents} Agents: {human_readable_formatted_value}") 
            metric_agent_count_values_for_latex_row.append(value_for_latex_cell) 
            
        latex_row_cells.extend(metric_agent_count_values_for_latex_row)

    latex_row_string = " & ".join(latex_row_cells) + " \\\\"
    current_method_console_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    current_method_file_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    
    for line in current_method_console_output_lines:
        print(line)
    print("-" * 30)
    
    all_methods_output_for_file.append("\n".join(current_method_file_output_lines))

# Optionally, write to a file
# with open("combined_25x25_warehouse_results_summary.txt", "w") as f:
#     f.write("\n\n".join(all_methods_output_for_file))
# print("\nSummary written to combined_25x25_warehouse_results_summary.txt")


```

## Static vs. Dynamic Replanning

```python
import os
import pickle
import numpy as np

# --- Configuration ---
DATA_DIR   = "simulation_data/results" # Ensure this path is correct

# Methods for this specific table data generation
methods_keys_for_table = ['bfs-dqn', 'bfs-ppo', 'astar-dqn', 'astar-ppo']
# Corresponding labels for the table output
method_labels_for_table_output = ['BFS-DQN', 'BFS-PPO', 'A*-DQN', 'A*-PPO']

# Agent counts for which data is loaded (individual agent numbers)
agent_counts_individual = np.arange(5, 21) # Corresponds to agents 5, 6, ..., 20

# Suffixes for filenames
suffix = {
    'bfs-dqn':   f"_do_bfs_dqn",
    'bfs-ppo':   f"_do_bfs_ppo",
    'astar-dqn': f"_do_astar_dqn",
    'astar-ppo': f"_do_astar_ppo",
}

# Mapping internal metric keys from files to the table's desired abbreviations
# The order here will also define the order in the output for metrics.
table_metric_map_ordered = {
    'success_rate': 'SR',
    'makespan': 'MS',
    'total_collisions': 'CO',
    'total_time_sec': 'T(s)'
}

# --- Data Loading and Processing Function (Consistent with previous table script) ---
def load_and_extract(path, metric_key, is_success_rate_metric, method_file_key):
    """
    Load data from a pickle file and extract the mean of a specified metric.
    - path: Path to the pickle file.
    - metric_key: The key for the metric in the data dictionary.
    - is_success_rate_metric: Boolean, True if the metric is 'success_rate'.
    - method_file_key: The key for the method, used for specific logic.
    """
    try:
        with open(path, 'rb') as f:
            runs_data = pickle.load(f)
    except FileNotFoundError:
        return np.nan
    
    metric_values_for_averaging = []
    if not runs_data: return np.nan

    for run_entry in runs_data:
        result_dict = run_entry.get('result', run_entry)
        if not isinstance(result_dict, dict): continue

        current_run_success_rate = result_dict.get('success_rate', 0.0)
        val_to_append = np.nan 

        if is_success_rate_metric:
            metric_values_for_averaging.append(1.0 if current_run_success_rate == 1.0 else 0.0)
            continue 

        if metric_key == 'total_time_sec': # Time is averaged over ALL runs
            val_to_append = result_dict.get('total_time_sec', np.nan)
        elif metric_key == 'makespan' or metric_key == 'total_collisions': # MS and CO over SOLVED runs
            if current_run_success_rate == 1.0:
                if metric_key == 'total_collisions':
                    # DQN/PPO methods likely use 'total_collisions'
                    if method_file_key in ['bfs-dqn', 'bfs-ppo', 'astar-dqn', 'astar-ppo']:
                        val_to_append = result_dict.get('total_collisions', np.nan)
                    else: # For other methods like CBS, ICBS (if they were included)
                        val_to_append = result_dict.get('unique_collisions', np.nan)
                elif metric_key == 'makespan':
                    val_to_append = result_dict.get('makespan', np.nan)
        
        if val_to_append is not None and not np.isnan(val_to_append):
            metric_values_for_averaging.append(float(val_to_append))

    if not metric_values_for_averaging:
        return np.nan
    return np.nanmean(metric_values_for_averaging)

# --- Load average metrics for each individual agent count ---
avg_metrics_data_individual = {method_key: {} for method_key in methods_keys_for_table}
print("Loading data for individual agent counts (for table)...")
for method_idx, method_file_key in enumerate(methods_keys_for_table):
    print(f"  Loading for method: {method_labels_for_table_output[method_idx]} ({method_file_key})")
    for metric_key_in_file in table_metric_map_ordered.keys():
        is_success_rate_metric = (metric_key_in_file == 'success_rate')
        metric_averages_for_method = []
        for num_agents in agent_counts_individual:
            filename_suffix_key = method_file_key # Key to look up in suffix dict
            if filename_suffix_key not in suffix:
                print(f"    Warning: Suffix for {filename_suffix_key} not found. Cannot generate filename for {num_agents} agents.")
                avg_val = np.nan
            else:
                filename = os.path.join(DATA_DIR, f"stat_simulation_results{suffix[filename_suffix_key]}_{num_agents}.pkl")
                avg_val = load_and_extract(filename, metric_key_in_file, is_success_rate_metric, method_file_key)
            metric_averages_for_method.append(avg_val)
        
        processed_averages = np.array(metric_averages_for_method, dtype=float)
        
        if metric_key_in_file == 'total_collisions': 
            processed_averages[processed_averages <= 0] = np.nan
            
        avg_metrics_data_individual[method_file_key][metric_key_in_file] = processed_averages
print("Data loading for table complete.\n")

# --- Define Agent Groups for Table (e.g., 11x11 style) ---
agent_groups_for_table_ordered = {
    "5-10": (5, 10),
    "11-15": (11, 15),
    "16-20": (16, 20)
}

# --- Calculate Averages for Agent Groups ---
table_data_new_methods = {
    method_label: {
        metric_abbr: {grp_label: np.nan for grp_label in agent_groups_for_table_ordered}
        for metric_abbr in table_metric_map_ordered.values()
    }
    for method_label in method_labels_for_table_output
}
min_agent_from_config = agent_counts_individual[0]

for method_idx, method_file_key in enumerate(methods_keys_for_table):
    current_method_label = method_labels_for_table_output[method_idx]
    for metric_key_in_file, table_metric_abbr in table_metric_map_ordered.items():
        # Ensure the metric_key_in_file was loaded for the method
        if metric_key_in_file not in avg_metrics_data_individual[method_file_key]:
             print(f"Warning: Metric key {metric_key_in_file} not found in loaded data for {method_file_key}. Skipping for table.")
             continue
        individual_averages = avg_metrics_data_individual[method_file_key][metric_key_in_file]

        for group_label, (start_agent, end_agent) in agent_groups_for_table_ordered.items():
            start_idx = start_agent - min_agent_from_config
            end_idx = end_agent - min_agent_from_config + 1 

            if start_idx < 0 or end_idx > len(individual_averages) or start_idx >= end_idx :
                print(f"Warning: Agent group {group_label} ({start_agent}-{end_agent}) has invalid slice [{start_idx}:{end_idx}] for loaded data.")
                table_data_new_methods[current_method_label][table_metric_abbr][group_label] = np.nan
                continue
                
            group_values = individual_averages[start_idx:end_idx]
            mean_for_group = np.nanmean(group_values) 
            table_data_new_methods[current_method_label][table_metric_abbr][group_label] = mean_for_group

# --- Format and Print Results for New Methods Table ---
print("--- Results for New Methods Table (BFS-DQN/PPO, A*-DQN/PPO) ---")
all_methods_output_for_file = []

all_methods_output_for_file.append(f"Data Presentation Note:")
all_methods_output_for_file.append(f"  - Success Rate (SR) is averaged over all trials.")
all_methods_output_for_file.append(f"  - Time (T(s)) is averaged over ALL trials (reflecting timeouts).")
all_methods_output_for_file.append(f"  - Makespan (MS) and Collisions (CO) are averaged *only* over successful trials.")
all_methods_output_for_file.append(f"  - A '-' for MS/CO indicates no successful trials in group with valid data.")
all_methods_output_for_file.append(f"  - A '-' for T(s) indicates no trials in group provided valid time data.\n")

all_methods_output_for_file.append(f"LaTeX Table Column Structure Hint (Example for 11x11 style):")
all_methods_output_for_file.append(f"  Metrics (super-columns): {list(table_metric_map_ordered.values())}")
all_methods_output_for_file.append(f"  Agent Groups per Metric (sub-columns): {list(agent_groups_for_table_ordered.keys())}\n")

for method_label in method_labels_for_table_output:
    current_method_console_output_lines = [f"Method: {method_label}"]
    current_method_file_output_lines = [f"Method: {method_label}"] 
    
    latex_row_cells = [method_label] 

    for metric_abbr in table_metric_map_ordered.values(): 
        current_method_console_output_lines.append(f"  {metric_abbr}:")
        current_method_file_output_lines.append(f"  {metric_abbr}:")
        
        metric_group_values_for_latex_row = [] 

        for group_label in agent_groups_for_table_ordered:
            raw_value = table_data_new_methods[method_label][metric_abbr][group_label]
            
            human_readable_formatted_value = "-" 
            if not np.isnan(raw_value):
                if metric_abbr == 'SR':
                    human_readable_formatted_value = f"{raw_value * 100:.2f}%"
                elif metric_abbr == 'CO':
                    human_readable_formatted_value = f"{raw_value:.0f}"
                else: # MS, T(s)
                    human_readable_formatted_value = f"{raw_value:.1f}"
            
            value_for_console_and_latex_cell = human_readable_formatted_value
            if metric_abbr == 'SR' and not np.isnan(raw_value): 
                value_for_console_and_latex_cell = human_readable_formatted_value.replace('%', '\\%') 
            
            current_method_console_output_lines.append(f"    {group_label}: {value_for_console_and_latex_cell}")
            current_method_file_output_lines.append(f"    {group_label}: {human_readable_formatted_value}") 
            metric_group_values_for_latex_row.append(value_for_console_and_latex_cell)
            
        latex_row_cells.extend(metric_group_values_for_latex_row)

    latex_row_string = " & ".join(latex_row_cells) + " \\\\"
    current_method_console_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    current_method_file_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    
    for line in current_method_console_output_lines:
        print(line)
    print("-" * 30)
    
    all_methods_output_for_file.append("\n".join(current_method_file_output_lines))

# --- Save results to a text file ---
output_filename = "table_data_new_methods_results.txt"
with open(output_filename, "w") as f:
    f.write("--- Simulation Data for New Methods Table (BFS-DQN/PPO, A*-DQN/PPO) ---\n")
    f.write("--- Time (T(s)) averaged over ALL trials ---\n\n")
    f.write("\n\n".join(all_methods_output_for_file)) 

print(f"\nFormatted results for new methods table saved to: {output_filename}")
```

```python
import os
import pickle
import numpy as np

# --- Configuration (Mostly Global) ---
# Methods for this specific table data generation
methods_keys_for_table = ['bfs-dqn', 'bfs-ppo', 'astar-dqn', 'astar-ppo']
# Corresponding labels for the table output
method_labels_for_table_output = ['BFS-DQN', 'BFS-PPO', 'A*-DQN', 'A*-PPO']

# Suffixes for filenames (method part)
suffix = {
    'bfs-dqn':   f"_do_bfs_dqn",
    'bfs-ppo':   f"_do_bfs_ppo",
    'astar-dqn': f"_do_astar_dqn",
    'astar-ppo': f"_do_astar_ppo",
}

# Mapping internal metric keys from files to the table's desired abbreviations
# The order here will also define the order in the output for metrics.
table_metric_map_ordered = {
    'success_rate': 'SR',
    'makespan': 'MS',
    'total_collisions': 'CO',
    'total_time_sec': 'T(s)'
}

# --- Scenario Definitions ---
scenarios = [
    {
        "name": "11x11 maze (5-20 agents)",
        "map_identifier_for_filename": "", # No specific infix for these files
        "base_data_dir": "simulation_data/results_model_comp",
        "agent_counts_individual": np.arange(5, 21), # Corresponds to agents 5, 6, ..., 20
        "agent_groups_for_table_ordered": { # Defines columns in the output table for this scenario
            "5-10": (5, 10),
            "11-15": (11, 15),
            "16-20": (16, 20)
        },
        "table_caption_detail": "For an 11x11 grid (5-20 agents). Example: Subtable (a)",
        "output_file_section_header": "--- Results for 11x11 Maze (Agents 5-20) ---"
    },
    {
        "name": "21x21 maze-like (32, 64, 96 agents)",
        "map_identifier_for_filename": "", # No specific infix
        "base_data_dir": "simulation_data/results_model_comp",
        "agent_counts_individual": np.array([32, 64, 96]),
        "agent_groups_for_table_ordered": {
            "32": (32, 32), # Each "group" is a single agent count
            "64": (64, 64),
            "96": (96, 96)
        },
        "table_caption_detail": "For a 21x21 grid-like env (32, 64, 96 agents). Example: Subtable (b)",
        "output_file_section_header": "--- Results for 21x21 Maze-like (Agents 32, 64, 96) ---"
    },
    {
        "name": "25x25 warehouse (32, 64, 96 agents)",
        "map_identifier_for_filename": "_wr", # Files have "_wr" in their names
        "base_data_dir": "simulation_data/results_model_comp",
        "agent_counts_individual": np.array([32, 64, 96]),
        "agent_groups_for_table_ordered": {
            "32": (32, 32),
            "64": (64, 64),
            "96": (96, 96)
        },
        "table_caption_detail": "For a 25x25 warehouse env (32, 64, 96 agents). Example: Subtable (c)",
        "output_file_section_header": "--- Results for 25x25 Warehouse (Agents 32, 64, 96) ---"
    }
]

# --- Data Loading and Processing Function (Consistent with previous table script) ---
def load_and_extract(path, metric_key, is_success_rate_metric, method_file_key):
    """
    Load data from a pickle file and extract the mean of a specified metric.
    - path: Path to the pickle file.
    - metric_key: The key for the metric in the data dictionary.
    - is_success_rate_metric: Boolean, True if the metric is 'success_rate'.
    - method_file_key: The key for the method, used for specific logic.
    """
    try:
        with open(path, 'rb') as f:
            runs_data = pickle.load(f)
    except FileNotFoundError:
        # print(f"    File not found: {path}") # Optional: for debugging missing files
        return np.nan
    
    metric_values_for_averaging = []
    if not runs_data: return np.nan

    for run_entry in runs_data:
        result_dict = run_entry.get('result', run_entry)
        if not isinstance(result_dict, dict): continue

        current_run_success_rate = result_dict.get('success_rate', 0.0)
        val_to_append = np.nan 

        if is_success_rate_metric:
            metric_values_for_averaging.append(1.0 if current_run_success_rate == 1.0 else 0.0)
            continue 

        if metric_key == 'total_time_sec': # Time is averaged over ALL runs
            val_to_append = result_dict.get('total_time_sec', np.nan)
        elif metric_key == 'makespan' or metric_key == 'total_collisions': # MS and CO over SOLVED runs
            if current_run_success_rate == 1.0:
                if metric_key == 'total_collisions':
                    # DQN/PPO methods likely use 'total_collisions'
                    if method_file_key in ['bfs-dqn', 'bfs-ppo', 'astar-dqn', 'astar-ppo']:
                        val_to_append = result_dict.get('total_collisions', np.nan)
                    else: # For other methods like CBS, ICBS (if they were included)
                        val_to_append = result_dict.get('unique_collisions', np.nan) # Should not be hit with current methods
                elif metric_key == 'makespan':
                    val_to_append = result_dict.get('makespan', np.nan)
        
        if val_to_append is not None and not np.isnan(val_to_append):
            metric_values_for_averaging.append(float(val_to_append))

    if not metric_values_for_averaging:
        return np.nan
    return np.nanmean(metric_values_for_averaging)

# --- Main Processing Loop for Scenarios ---
collated_output_for_master_file = []

# Add global data presentation notes once
collated_output_for_master_file.append(f"Data Presentation Note:")
collated_output_for_master_file.append(f"  - Success Rate (SR) is averaged over all trials.")
collated_output_for_master_file.append(f"  - Time (T(s)) is averaged over ALL trials (reflecting timeouts).")
collated_output_for_master_file.append(f"  - Makespan (MS) and Collisions (CO) are averaged *only* over successful trials.")
collated_output_for_master_file.append(f"  - A '-' for MS/CO indicates no successful trials in group with valid data.")
collated_output_for_master_file.append(f"  - A '-' for T(s) indicates no trials in group provided valid time data.\n")


for scenario in scenarios:
    print(f"\nProcessing Scenario: {scenario['name']}")
    collated_output_for_master_file.append(f"\n{scenario['output_file_section_header']}")
    collated_output_for_master_file.append(f"Table Caption Hint: {scenario['table_caption_detail']}")

    DATA_DIR = scenario['base_data_dir']
    current_agent_counts_individual = scenario['agent_counts_individual']
    current_agent_groups_for_table = scenario['agent_groups_for_table_ordered']
    map_id_for_filename = scenario['map_identifier_for_filename']

    # --- Load average metrics for each individual agent count for the current scenario ---
    avg_metrics_data_individual = {method_key: {} for method_key in methods_keys_for_table}
    print(f"  Loading data for individual agent counts ({scenario['name']})...")
    for method_idx, method_file_key in enumerate(methods_keys_for_table):
        print(f"    Loading for method: {method_labels_for_table_output[method_idx]} ({method_file_key})")
        for metric_key_in_file in table_metric_map_ordered.keys():
            is_success_rate_metric = (metric_key_in_file == 'success_rate')
            metric_averages_for_method = []
            for num_agents in current_agent_counts_individual:
                filename_suffix_key = method_file_key 
                if filename_suffix_key not in suffix:
                    print(f"      Warning: Suffix for {filename_suffix_key} not found. Cannot generate filename for {num_agents} agents.")
                    avg_val = np.nan
                else:
                    # Construct filename using map_id_for_filename
                    filename = os.path.join(DATA_DIR, f"simulation_results{map_id_for_filename}{suffix[filename_suffix_key]}_{num_agents}.pkl")
                    avg_val = load_and_extract(filename, metric_key_in_file, is_success_rate_metric, method_file_key)
                metric_averages_for_method.append(avg_val)
            
            processed_averages = np.array(metric_averages_for_method, dtype=float)
            
            if metric_key_in_file == 'total_collisions': 
                # Ensure 0 collisions from successful runs without collisions are not turned to NaN,
                # but if it was NaN (e.g. no successful runs) it stays NaN.
                # This might need adjustment based on how 0 collisions are recorded vs. no data.
                # The original code `processed_averages[processed_averages <= 0] = np.nan` would make actual 0s into NaN.
                # Let's assume 0 is a valid collision count if successful.
                # If the goal is to show '-' when no successful runs, load_and_extract already returns NaN for that case.
                pass # Keep as is, or refine if 0 values need specific NaN conversion beyond what load_and_extract does.
                
            avg_metrics_data_individual[method_file_key][metric_key_in_file] = processed_averages
    print(f"  Data loading for scenario '{scenario['name']}' complete.\n")

    # --- Calculate Averages for Agent Groups for the current scenario ---
    table_data_current_scenario = {
        method_label: {
            metric_abbr: {grp_label: np.nan for grp_label in current_agent_groups_for_table}
            for metric_abbr in table_metric_map_ordered.values()
        }
        for method_label in method_labels_for_table_output
    }

    for method_idx, method_file_key in enumerate(methods_keys_for_table):
        current_method_label = method_labels_for_table_output[method_idx]
        for metric_key_in_file, table_metric_abbr in table_metric_map_ordered.items():
            if metric_key_in_file not in avg_metrics_data_individual[method_file_key]:
                 print(f"    Warning: Metric key {metric_key_in_file} not found in loaded data for {method_file_key} in scenario {scenario['name']}. Skipping for table.")
                 continue
            
            # This array holds averages for individual agent counts (e.g., for 5, 6,...20 or for 32, 64, 96)
            individual_averages_for_metric = avg_metrics_data_individual[method_file_key][metric_key_in_file]

            for group_label, (start_agent, end_agent) in current_agent_groups_for_table.items():
                # Find indices in current_agent_counts_individual that fall within this group's agent range
                agent_indices_in_loaded_data = [
                    i for i, ac_val in enumerate(current_agent_counts_individual)
                    if start_agent <= ac_val <= end_agent
                ]

                if not agent_indices_in_loaded_data:
                    # This case means the defined group (e.g., "5-10 agents") has no corresponding
                    # agent numbers in the `current_agent_counts_individual` for this scenario.
                    # This shouldn't happen if `current_agent_groups_for_table` is defined correctly
                    # relative to `current_agent_counts_individual`.
                    print(f"      Warning: Agent group {group_label} ({start_agent}-{end_agent}) found no matching agent counts in {current_agent_counts_individual} for scenario '{scenario['name']}'.")
                    values_for_group_mean = np.array([])
                else:
                    values_for_group_mean = individual_averages_for_metric[agent_indices_in_loaded_data]
                
                mean_for_group = np.nanmean(values_for_group_mean) 
                table_data_current_scenario[current_method_label][table_metric_abbr][group_label] = mean_for_group
    
    # --- Format and Print/Store Results for the Current Scenario ---
    scenario_specific_output = []
    
    scenario_specific_output.append(f"LaTeX Table Column Structure Hint:")
    scenario_specific_output.append(f"  Metrics (super-columns): {list(table_metric_map_ordered.values())}")
    scenario_specific_output.append(f"  Agent Groups per Metric (sub-columns): {list(current_agent_groups_for_table.keys())}\n")

    for method_label in method_labels_for_table_output:
        current_method_console_output_lines = [f"Method: {method_label}"]
        # current_method_file_output_lines will be built for scenario_specific_output
        
        latex_row_cells = [method_label] 

        for metric_abbr in table_metric_map_ordered.values(): 
            current_method_console_output_lines.append(f"  {metric_abbr}:")
            
            metric_group_values_for_latex_row = [] 

            for group_label in current_agent_groups_for_table: # Use current scenario's groups
                raw_value = table_data_current_scenario[method_label][metric_abbr][group_label]
                
                human_readable_formatted_value = "-" 
                if not np.isnan(raw_value):
                    if metric_abbr == 'SR':
                        human_readable_formatted_value = f"{raw_value * 100:.2f}%"
                    elif metric_abbr == 'CO': # Collisions typically integers
                        human_readable_formatted_value = f"{raw_value:.0f}"
                    else: # MS, T(s)
                        human_readable_formatted_value = f"{raw_value:.1f}"
                
                value_for_console_and_latex_cell = human_readable_formatted_value
                if metric_abbr == 'SR' and not np.isnan(raw_value): 
                    # Escape % for LaTeX
                    value_for_console_and_latex_cell = human_readable_formatted_value.replace('%', '\\%') 
                
                current_method_console_output_lines.append(f"    {group_label}: {value_for_console_and_latex_cell}")
                metric_group_values_for_latex_row.append(value_for_console_and_latex_cell)
                
            latex_row_cells.extend(metric_group_values_for_latex_row)

        latex_row_string = " & ".join(latex_row_cells) + " \\\\"
        current_method_console_output_lines.append(f"  LaTeX row data: {latex_row_string}")
        
        # Print to console for immediate feedback
        for line in current_method_console_output_lines:
            print(line)
        print("-" * 30)
        
        # Store for file output (rebuilding a clean version for the file)
        file_lines_for_method = [f"Method: {method_label}"]
        for metric_abbr_file in table_metric_map_ordered.values():
            file_lines_for_method.append(f"  {metric_abbr_file}:")
            for group_label_file in current_agent_groups_for_table:
                # Re-fetch and re-format for clarity, ensuring consistency
                raw_val_file = table_data_current_scenario[method_label][metric_abbr_file][group_label_file]
                hr_val_file = "-"
                if not np.isnan(raw_val_file):
                    if metric_abbr_file == 'SR': hr_val_file = f"{raw_val_file * 100:.2f}%"
                    elif metric_abbr_file == 'CO': hr_val_file = f"{raw_val_file:.0f}"
                    else: hr_val_file = f"{raw_val_file:.1f}"
                file_lines_for_method.append(f"    {group_label_file}: {hr_val_file}") # Non-escaped for .txt file
        file_lines_for_method.append(f"  LaTeX row data: {latex_row_string}")
        scenario_specific_output.append("\n".join(file_lines_for_method))

    collated_output_for_master_file.append("\n\n".join(scenario_specific_output))
    collated_output_for_master_file.append("-" * 60) # Separator between scenarios in file

# --- Save combined results to a text file ---
output_filename = "simulation_data/results_model_comp/collated_table_data_results.txt"
with open(output_filename, "w") as f:
    f.write("--- COLLATED SIMULATION DATA FOR LATEX TABLES (BFS-DQN/PPO, A*-DQN/PPO) ---\n")
    # Global notes already added at the start of collated_output_for_master_file
    f.write("\n\n".join(collated_output_for_master_file)) 

print(f"\nFormatted and collated results for all scenarios saved to: {output_filename}")
```

```python
import os
import pickle
import numpy as np

# --- Configuration ---
# Methods to process (BFS-DQN and A*-DQN)
methods_keys_for_table = ['bfs-dqn', 'astar-dqn']
method_labels_for_table_output = ['Our-BFS', 'Our-A*'] # Labels for the output

# Suffixes for filenames (method part)
suffix = {
    'bfs-dqn':   f"_do_bfs_dqn",
    'astar-dqn': f"_do_astar_dqn",
}

# Mapping internal metric keys from files to the table's desired abbreviations
table_metric_map_ordered = {
    'success_rate': 'SR',
    'makespan': 'MS',
    'total_collisions': 'CO',
    'total_time_sec': 'T(s)'
}

# --- Agent Selection Policy Definitions ---
agent_selection_policies = [
    {
        "name": "Random Agent Selection Policy",
        "filename_infix": "_rand", # Added after method suffix, before agent number
        "latex_multicolumn_header": "\\multicolumn{13}{c}{\\textit{Policy: Random Agent Selection}} \\\\ \\midrule"
    },
    {
        "name": "Farthest Agent Selection Policy",
        "filename_infix": "_far",  # Added after method suffix, before agent number
        "latex_multicolumn_header": "\\multicolumn{13}{c}{\\textit{Policy: Farthest Agent Selection}} \\\\ \\midrule"
    },
    {
        "name": "Fewest Future Collisions (FFC) Policy",
        "filename_infix": "",  # No infix for FFC (default)
        "latex_multicolumn_header": "\\multicolumn{13}{c}{\\textit{Policy: Fewest Future Collisions (FFC)}} \\\\ \\midrule"
    }
]

# --- Scenario (Map Configuration) Definitions ---
scenarios = [
    {
        "name": "11x11 maze (5-20 agents) - Agent Policy Comparison",
        "map_identifier_for_filename": "", # No specific map infix for base 11x11
        "base_data_dir": "simulation_data/results_agent_policy", # New data directory
        "agent_counts_individual": np.arange(5, 21),
        "agent_groups_for_table_ordered": {
            "5-10": (5, 10), "11-15": (11, 15), "16-20": (16, 20)
        },
        "output_file_section_header": "--- Results for Agent Policy Comparison: 11x11 Maze (Agents 5-20) ---",
        "table_caption_detail": "For an 11x11 grid (5-20 agents), comparing agent selection policies. Time in seconds."
    },
    {
        "name": "21x21 maze-like (32, 64, 96 agents) - Agent Policy Comparison",
        "map_identifier_for_filename": "",
        "base_data_dir": "simulation_data/results_agent_policy",
        "agent_counts_individual": np.array([32, 64, 96]),
        "agent_groups_for_table_ordered": {
            "32": (32, 32), "64": (64, 64), "96": (96, 96)
        },
        "output_file_section_header": "--- Results for Agent Policy Comparison: 21x21 Maze-like (Agents 32, 64, 96) ---",
        "table_caption_detail": "For a 21x21 grid-like env (32, 64, 96 agents), comparing agent selection policies. Time in seconds."
    },
    {
        "name": "25x25 warehouse (32, 64, 96 agents) - Agent Policy Comparison",
        "map_identifier_for_filename": "_wr", # Warehouse files have "_wr" infix
        "base_data_dir": "simulation_data/results_agent_policy",
        "agent_counts_individual": np.array([32, 64, 96]),
        "agent_groups_for_table_ordered": {
            "32": (32, 32), "64": (64, 64), "96": (96, 96)
        },
        "output_file_section_header": "--- Results for Agent Policy Comparison: 25x25 Warehouse (Agents 32, 64, 96) ---",
        "table_caption_detail": "For a 25x25 warehouse env (32, 64, 96 agents), comparing agent selection policies. Time in seconds."
    }
]

# --- Data Loading and Processing Function (Reusable) ---
def load_and_extract(path, metric_key, is_success_rate_metric, method_file_key):
    try:
        with open(path, 'rb') as f:
            runs_data = pickle.load(f)
    except FileNotFoundError:
        # print(f"    File not found: {path}") # Optional for debugging
        return np.nan
    
    metric_values_for_averaging = []
    if not runs_data: return np.nan

    for run_entry in runs_data:
        result_dict = run_entry.get('result', run_entry)
        if not isinstance(result_dict, dict): continue

        current_run_success_rate = result_dict.get('success_rate', 0.0)
        val_to_append = np.nan 

        if is_success_rate_metric:
            metric_values_for_averaging.append(1.0 if current_run_success_rate == 1.0 else 0.0)
            continue 

        if metric_key == 'total_time_sec':
            val_to_append = result_dict.get('total_time_sec', np.nan)
        elif metric_key == 'makespan' or metric_key == 'total_collisions':
            if current_run_success_rate == 1.0: # Only for successful runs
                val_to_append = result_dict.get(metric_key, np.nan)
        
        if val_to_append is not None and not np.isnan(val_to_append):
            metric_values_for_averaging.append(float(val_to_append))

    if not metric_values_for_averaging:
        return np.nan
    return np.nanmean(metric_values_for_averaging)

# --- Main Processing Loop ---
collated_output_for_master_file = []

collated_output_for_master_file.append(f"Data Presentation Note:")
collated_output_for_master_file.append(f"  - Success Rate (SR) is averaged over all trials.")
collated_output_for_master_file.append(f"  - Time (T(s)) is averaged over ALL trials (reflecting timeouts).")
collated_output_for_master_file.append(f"  - Makespan (MS) and Collisions (CO) are averaged *only* over successful trials.")
collated_output_for_master_file.append(f"  - A '-' for MS/CO indicates no successful trials in group with valid data for that metric.")
collated_output_for_master_file.append(f"  - A '-' for T(s) indicates no trials in group provided valid time data.\n")

for scenario in scenarios:
    print(f"\nProcessing Scenario: {scenario['name']}")
    collated_output_for_master_file.append(f"\n{scenario['output_file_section_header']}")
    collated_output_for_master_file.append(f"Table Caption Hint: {scenario['table_caption_detail']}")
    collated_output_for_master_file.append(f"LaTeX Table Column Structure Hint:")
    collated_output_for_master_file.append(f"  Metrics (super-columns): {list(table_metric_map_ordered.values())}")
    collated_output_for_master_file.append(f"  Agent Groups per Metric (sub-columns): {list(scenario['agent_groups_for_table_ordered'].keys())}\n")

    scenario_base_data_dir = scenario['base_data_dir']
    current_agent_counts_individual = scenario['agent_counts_individual']
    current_agent_groups_for_table = scenario['agent_groups_for_table_ordered']
    map_id_for_filename = scenario['map_identifier_for_filename']
    
    lines_for_current_scenario_output_file = []

    for policy in agent_selection_policies: # Changed from 'setting' to 'policy'
        print(f"  Processing Policy: {policy['name']}")
        lines_for_current_scenario_output_file.append(f"{policy['latex_multicolumn_header']}")

        # Load average metrics for this policy and scenario
        avg_metrics_data_individual = {m_key: {} for m_key in methods_keys_for_table}
        for method_idx, method_file_key in enumerate(methods_keys_for_table):
            for metric_key_in_file in table_metric_map_ordered.keys():
                is_success_rate = (metric_key_in_file == 'success_rate')
                metric_averages_for_method = []
                for num_agents in current_agent_counts_individual:
                    method_filename_part = suffix[method_file_key] # e.g., "_do_bfs_dqn"
                    policy_infix_part = policy['filename_infix']   # e.g., "_rand", "_far", or ""
                    
                    filename = os.path.join(
                        scenario_base_data_dir,
                        f"simulation_results{map_id_for_filename}{method_filename_part}{policy_infix_part}_{num_agents}.pkl"
                    )
                    avg_val = load_and_extract(filename, metric_key_in_file, is_success_rate, method_file_key)
                    metric_averages_for_method.append(avg_val)
                
                processed_averages = np.array(metric_averages_for_method, dtype=float)
                avg_metrics_data_individual[method_file_key][metric_key_in_file] = processed_averages
        
        # Calculate Averages for Agent Groups for this policy
        table_data_for_policy = { # Changed from 'table_data_for_setting'
            m_label: {
                metric_abbr: {grp_label: np.nan for grp_label in current_agent_groups_for_table}
                for metric_abbr in table_metric_map_ordered.values()
            } for m_label in method_labels_for_table_output
        }

        for method_idx, method_file_key in enumerate(methods_keys_for_table):
            current_method_label = method_labels_for_table_output[method_idx]
            for metric_key_in_file, table_metric_abbr in table_metric_map_ordered.items():
                if metric_key_in_file not in avg_metrics_data_individual[method_file_key]:
                    # This might happen if a file for a specific method under a policy is missing
                    print(f"      Warning: Data for {metric_key_in_file} not found for method {method_file_key} under policy '{policy['name']}' in scenario '{scenario['name']}'.")
                    continue
                
                individual_averages = avg_metrics_data_individual[method_file_key][metric_key_in_file]
                for group_label, (start_agent, end_agent) in current_agent_groups_for_table.items():
                    indices = [i for i, ac in enumerate(current_agent_counts_individual) if start_agent <= ac <= end_agent]
                    group_values = individual_averages[indices] if indices else np.array([])
                    mean_for_group = np.nanmean(group_values) if group_values.size > 0 else np.nan
                    table_data_for_policy[current_method_label][table_metric_abbr][group_label] = mean_for_group

        # Format and Store Results for this policy
        for method_label in method_labels_for_table_output:
            latex_row_cells = [method_label]
            
            for metric_abbr in table_metric_map_ordered.values():
                metric_group_values_for_latex = []
                for group_label in current_agent_groups_for_table:
                    raw_value = table_data_for_policy[method_label][metric_abbr][group_label]
                    formatted_val = "-"
                    if not np.isnan(raw_value):
                        if metric_abbr == 'SR': formatted_val = f"{raw_value * 100:.2f}\\%"
                        elif metric_abbr == 'CO': formatted_val = f"{raw_value:.0f}"
                        else: formatted_val = f"{raw_value:.1f}" # MS, T(s)
                    metric_group_values_for_latex.append(formatted_val)
                latex_row_cells.extend(metric_group_values_for_latex)
            
            latex_row_string = " & ".join(latex_row_cells) + " \\\\"
            lines_for_current_scenario_output_file.append(f"Method: {method_label}")
            lines_for_current_scenario_output_file.append(f"  LaTeX row data: {latex_row_string}")
        lines_for_current_scenario_output_file.append("") # Blank line between policies' method blocks

    collated_output_for_master_file.append("\n".join(lines_for_current_scenario_output_file))
    collated_output_for_master_file.append("-" * 60 + "\n") # Scenario separator

# --- Save combined results to a text file ---
output_filename = "simulation_data/results_agent_policy/collated_table_data_agent_policy_results.txt"
with open(output_filename, "w") as f:
    f.write("--- COLLATED SIMULATION DATA FOR AGENT POLICY COMPARISON LATEX TABLES ---\n")
    f.write("\n".join(collated_output_for_master_file))

print(f"\nFormatted and collated results for agent policy comparison saved to: {output_filename}")
```

## Random Map Results

```python
import os
import pickle
import numpy as np

# --- Configuration for 32x32 Random Map ---
DATA_DIR = "simulation_data/results_random"  # Updated path

# Methods for this specific table data generation
methods_keys_for_table = ['bfs-dqn', 'astar-dqn', 'scrimp', 'primal', 'cbs', 'icbs']
# Corresponding labels for the table output
method_labels_for_table_output = ['Our-BFS-DQN', 'Our-A*-DQN', 'SCRIMP', 'PRIMAL', 'CBS', 'ICBS']

# Agent counts to process for the 32x32 random grid
agent_counts_to_process = [8, 16, 32, 64, 96]

# Suffixes for filenames - these correspond to the part of the filename
# AFTER "random" and BEFORE "_{num_agents}.pkl"
# Example: random_CBS_8.pkl -> suffix is _CBS
# Example: random_results_do_astar_dqn_8.pkl -> suffix is _results_do_astar_dqn
file_name_parts = {
    'bfs-dqn':   "_results_do_bfs_dqn",
    'astar-dqn': "_results_do_astar_dqn",
    'scrimp':    "_scrimp",
    'primal':    "_primal",
    'cbs':       "_CBS",
    'icbs':      "_ICBS",
}

# Mapping internal metric keys from files to the table's desired abbreviations
table_metric_map_ordered = {
    'success_rate': 'SR',
    'makespan': 'MS',
    'total_collisions': 'CO',
    'total_time_sec': 'T(s)',
    'episode_len': 'EpLen' # Internal key for episode length
}

# Placeholder for unsolved PRIMAL episode length on 32x32 grid
# VERIFY THIS VALUE: It's often a timeout or max_steps (e.g., 32*32=1024, or a fixed experimental value)
PRIMAL_UNSOLVED_EPLEN_32X32 = 1024.0

# --- Data Loading and Processing Function (UPDATED) ---
def load_and_extract(path, metric_key, is_success_rate_metric, method_file_key):
    """
    Load data from a pickle file and extract the mean of a specified metric.
    - path: Path to the pickle file.
    - metric_key: The key for the metric in the data dictionary.
    - is_success_rate_metric: Boolean, True if the metric is 'success_rate'.
    - method_file_key: The key for the method, used for specific logic.
    """
    try:
        with open(path, 'rb') as f:
            runs_data = pickle.load(f)
    except FileNotFoundError:
        # print(f"Warning: File not found {path}")
        return np.nan
    
    metric_values_for_averaging = []
    if not runs_data: return np.nan

    for run_entry in runs_data:
        # Accommodate if 'result' is nested or if run_entry is the result dict itself
        result_dict = run_entry.get('result', run_entry if isinstance(run_entry, dict) else {})
        if not isinstance(result_dict, dict): continue

        current_run_success_rate = result_dict.get('success_rate', 0.0)
        val_to_append = np.nan 

        if is_success_rate_metric: # This is 'success_rate'
            metric_values_for_averaging.append(1.0 if current_run_success_rate == 1.0 else 0.0)
            continue 

        # For other metrics
        if metric_key == 'total_time_sec': # Time is averaged over ALL runs
            val_to_append = result_dict.get('total_time_sec', np.nan)
        elif metric_key == 'episode_len': # Episode Length handling
            if method_file_key == 'primal':
                if current_run_success_rate == 1.0:
                    val_to_append = result_dict.get('makespan', np.nan) # EpLen is Makespan for solved PRIMAL
                else:
                    val_to_append = PRIMAL_UNSOLVED_EPLEN_32X32 # Updated for 32x32 grid
            else: # For SCRIMP, CBS, ICBS, and other methods (BFS-DQN, A*-DQN, etc.)
                val_to_append = result_dict.get('episode_len', np.nan) # Averaged over ALL runs
        elif metric_key == 'makespan' or metric_key == 'total_collisions': # MS and CO over SOLVED runs
            if current_run_success_rate == 1.0:
                if metric_key == 'total_collisions':
                    # Logic for collisions based on method type
                    if method_file_key in ['bfs-dqn', 'astar-dqn', 'bfs-ppo', 'astar-ppo', 'scrimp']:
                        val_to_append = result_dict.get('total_collisions', np.nan)
                    else: # For PRIMAL, CBS, ICBS if they report 'unique_collisions'
                        val_to_append = result_dict.get('unique_collisions', np.nan)
                elif metric_key == 'makespan':
                    val_to_append = result_dict.get('makespan', np.nan)
        
        if val_to_append is not None and not np.isnan(val_to_append):
            metric_values_for_averaging.append(float(val_to_append))

    if not metric_values_for_averaging:
        return np.nan
    return np.nanmean(metric_values_for_averaging)

# --- Load average metrics for each specified agent count ---
results_data_32x32_random = {
    method_key: {
        metric_key_in_file: {agent_count: np.nan for agent_count in agent_counts_to_process}
        for metric_key_in_file in table_metric_map_ordered.keys() 
    }
    for method_key in methods_keys_for_table
}

print(f"Loading data for 32x32 Random Map (Agents: {agent_counts_to_process})...")
for method_idx, method_file_key in enumerate(methods_keys_for_table):
    print(f"  Loading for method: {method_labels_for_table_output[method_idx]} ({method_file_key})")
    if method_file_key not in file_name_parts:
        print(f"    Warning: File name part for '{method_file_key}' not found in 'file_name_parts'. Skipping this method.")
        # Initialize with NaNs if file_name_part is missing
        for metric_key_in_file in table_metric_map_ordered.keys():
            for num_agents in agent_counts_to_process:
                results_data_32x32_random[method_file_key][metric_key_in_file][num_agents] = np.nan
        continue

    current_file_name_part = file_name_parts[method_file_key]
    for metric_key_in_file in table_metric_map_ordered.keys(): 
        is_success_rate_metric = (metric_key_in_file == 'success_rate')
        for num_agents in agent_counts_to_process:
            # Construct filename based on the new pattern
            filename = os.path.join(DATA_DIR, f"random{current_file_name_part}_{num_agents}.pkl")
            avg_val = load_and_extract(filename, metric_key_in_file, is_success_rate_metric, method_file_key)
            
            # If total_collisions is 0 or less (and not NaN), treat as NaN (as per original logic)
            if metric_key_in_file == 'total_collisions' and avg_val is not None and avg_val <= 0:
                avg_val = np.nan 
            
            results_data_32x32_random[method_file_key][metric_key_in_file][num_agents] = avg_val
print("Data loading complete.\n")


# --- Format and Print Results for 32x32 Random Map Table ---
grid_description = "32x32 (Random Map)"
print(f"--- Results for {grid_description} Table (Methods: {', '.join(method_labels_for_table_output)}; Agents: {agent_counts_to_process}) ---")
all_methods_output_for_file = []

# Data Presentation Note
all_methods_output_for_file.append(f"Data Presentation Note:")
all_methods_output_for_file.append(f"  - Grid: {grid_description}")
all_methods_output_for_file.append(f"  - Agents: {agent_counts_to_process}")
all_methods_output_for_file.append(f"  - Success Rate (SR) is averaged over all trials.")
all_methods_output_for_file.append(f"  - Time (T(s)) and general Episode Length (EpLen) are averaged over ALL trials.")
all_methods_output_for_file.append(f"  - For PRIMAL, EpLen is Makespan for solved instances, {PRIMAL_UNSOLVED_EPLEN_32X32} for unsolved ({grid_description}).")
all_methods_output_for_file.append(f"  - For SCRIMP & PRIMAL, T(s) is shown as T(s) (Avg EpLen).")
all_methods_output_for_file.append(f"  - Makespan (MS) and Collisions (CO) are averaged *only* over successful trials.")
all_methods_output_for_file.append(f"  - A '-' for MS/CO indicates no successful trials with valid data for that metric.")
all_methods_output_for_file.append(f"  - A '-' for T(s)/EpLen indicates no trials provided valid data.\n")

header_metric_abbrs = [abbr for key, abbr in table_metric_map_ordered.items() if key != 'episode_len']
all_methods_output_for_file.append(f"LaTeX Table Column Structure Hint (Example for {grid_description} style):")
all_methods_output_for_file.append(f"  Metrics (super-columns): {header_metric_abbrs}")
all_methods_output_for_file.append(f"  Agent Counts per Metric (sub-columns): {agent_counts_to_process}\n")


for method_idx, method_file_key in enumerate(methods_keys_for_table):
    current_method_label = method_labels_for_table_output[method_idx]
    current_method_console_output_lines = [f"Method: {current_method_label}"]
    current_method_file_output_lines = [f"Method: {current_method_label}"]
    
    latex_row_cells = [current_method_label]

    for metric_key_in_file_display, metric_abbr_display in table_metric_map_ordered.items():
        if metric_key_in_file_display == 'episode_len': 
            continue # Skip direct printing of EpLen as it's combined for SCRIMP/PRIMAL Time

        current_method_console_output_lines.append(f"  {metric_abbr_display}:")
        current_method_file_output_lines.append(f"  {metric_abbr_display}:")
        
        metric_agent_count_values_for_latex_row = []

        for num_agents in agent_counts_to_process:
            raw_value = np.nan 
            if method_file_key in results_data_32x32_random and \
               metric_key_in_file_display in results_data_32x32_random[method_file_key]:
                raw_value = results_data_32x32_random[method_file_key][metric_key_in_file_display].get(num_agents, np.nan)
            else:
                print(f"    Warning: Data for {current_method_label} - {metric_abbr_display} - {num_agents} Agents not found. Using NaN.")

            human_readable_formatted_value = "-" 
            if not np.isnan(raw_value):
                if metric_abbr_display == 'SR':
                    human_readable_formatted_value = f"{raw_value * 100:.2f}%"
                elif metric_abbr_display == 'CO':
                    human_readable_formatted_value = f"{raw_value:.0f}" # Collisions as integer
                else: # MS, T(s)
                    human_readable_formatted_value = f"{raw_value:.1f}"
            
            # Special handling for SCRIMP and PRIMAL Time to include Episode Length
            if metric_key_in_file_display == 'total_time_sec' and \
               (method_file_key == 'scrimp' or method_file_key == 'primal'):
                avg_ep_len_for_agent_count = np.nan
                if method_file_key in results_data_32x32_random and \
                   'episode_len' in results_data_32x32_random[method_file_key]:
                     avg_ep_len_for_agent_count = results_data_32x32_random[method_file_key]['episode_len'].get(num_agents, np.nan)

                if not np.isnan(avg_ep_len_for_agent_count):
                    formatted_ep_len = f"{avg_ep_len_for_agent_count:.0f}" # EpLen as integer for display
                    if human_readable_formatted_value != "-":
                        human_readable_formatted_value += f" ({formatted_ep_len})"
                    else: 
                        # If time is NaN but EpLen is available
                        human_readable_formatted_value = f"- ({formatted_ep_len})"
            
            value_for_console_and_latex_cell = human_readable_formatted_value
            if metric_abbr_display == 'SR' and not np.isnan(raw_value): 
                value_for_console_and_latex_cell = human_readable_formatted_value.replace('%', '\\%') # Escape % for LaTeX
            
            current_method_console_output_lines.append(f"    {num_agents} Agents: {value_for_console_and_latex_cell}")
            current_method_file_output_lines.append(f"    {num_agents} Agents: {human_readable_formatted_value}") 
            metric_agent_count_values_for_latex_row.append(value_for_console_and_latex_cell)
            
        latex_row_cells.extend(metric_agent_count_values_for_latex_row)

    latex_row_string = " & ".join(latex_row_cells) + " \\\\"
    current_method_console_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    current_method_file_output_lines.append(f"  LaTeX row data: {latex_row_string}")
    
    for line in current_method_console_output_lines:
        print(line)
    print("-" * 30)
    
    all_methods_output_for_file.append("\n".join(current_method_file_output_lines))

# # --- Save results to a text file ---
# output_filename = f"table_data_32x32_random_map_agents_{'_'.join(map(str, agent_counts_to_process))}_eplen{int(PRIMAL_UNSOLVED_EPLEN_32X32)}.txt"
# with open(output_filename, "w") as f:
#     f.write(f"--- Simulation Data for {grid_description} (All Methods: {agent_counts_to_process} Agents) ---\n")
#     f.write(f"--- Time (T(s)) and EpLen averaged over ALL trials; EpLen details for PRIMAL/SCRIMP in T(s) column (PRIMAL unsolved EpLen={PRIMAL_UNSOLVED_EPLEN_32X32}) ---\n\n")
#     f.write("\n\n".join(all_methods_output_for_file)) 

# print(f"\nFormatted results saved to: {output_filename}")
```
